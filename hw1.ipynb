```python
x = [3,4,1,2,5,6,7,8,2,4]
h = [0.25, 0.25, 0.25]

def basic_convolution(x, h):
  h_reversed = h[::-1]
  x_padded = [0] * (len(h) - 1) + x + [0] * (len(h) - 1)
  output = []
  for index in range(len(x) + len(h) - 1):
    output.append(sum([x_padded[index:index + len(h)][i] * h_reversed[i] for i in range(len(h))]))
  return output

print(basic_convolution(x, h))
```

1b. 
This function takes the x list, pads it with zeroes so that each output of the convolution can add the 3 products of elements (or whatever the length of h is) from padded x and h, then append them to the output list. This involves a loop through the padded x list and then a sub-loop through the h list, leading to a time complexity of O(n^2).

1c.

```python
def basic_convolution_constant(x, h):
  h_constant = h[0]
  x_padded = [0] * (len(h) - 1) + x + [0] * (len(h) - 1)
  output = []
  for index in range(len(x) + len(h) - 1):
    output.append(sum(x_padded[index:index + len(h)]) * h_constant)
  return output

print(basic_convolution_constant(x, h))
```

1d.

```python
import time

x_long = x * 100
h_long = h * 100

start = time.time()
basic_convolution(x_long, h_long)
end1 = time.time()
a = basic_convolution_constant(x_long, h_long)
end2 = time.time()
print("O(n^2) time: " + str(end1 - start))
print("O(n) time: " + str(end2 - end1))
print(a[-5:])
```

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 1, 1000)
f = 2

def calculate_sum(k):
  total = np.zeros(1000)
  for val in range(1, k+1):
    total += np.sin(2 * np.pi * (2 * val - 1) * f * t) / (2 * val - 1)
  return total

ax1 = plt.subplot(511)
ax2 = plt.subplot(512)
ax3 = plt.subplot(513)
ax4 = plt.subplot(514)
ax5 = plt.subplot(515)

ax1.plot(calculate_sum(1))
ax1.title.set_text("K = 1")
ax2.plot(calculate_sum(3))
ax2.title.set_text("K = 3")
ax3.plot(calculate_sum(9))
ax3.title.set_text("K = 9")
ax4.plot(calculate_sum(50))
ax4.title.set_text("K = 50")
ax5.plot(calculate_sum(1000000))
ax5.title.set_text("K = 10e6")
plt.xlabel("t")
plt.ylabel("y(t)")
plt.show()
```
