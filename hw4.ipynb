```python
from scipy.io import wavfile
from scipy.signal import correlate
import numpy as np
import matplotlib.pyplot as plt
```

```python
def get_freqs(freq, N, file, freqs, start):
  max_f_val = 0
  x = wavfile.read('/content/gdrive/MyDrive/Cornell-Tech-Github/Digital-Signal-Processing/soundfiles/' + file)[1]
  for f in freqs:
    s1, s2 = 0, 0
    total = 0
    for n in range(N):
      s = x[start + n] + 2 * np.cos(2 * np.pi * f / freq) * s1 - s2
      power = s ** 2 + s1 ** 2 - np.cos(2 * np.pi * f / freq) * s * s1
      total += x[start + n] ** 2
      s1, s2 = s, s1
    out = power / total / N
    if out > max_f_val:
      max_f_val = out
      max_f = f
  return max_f

def get_key(freq, N, file, freqs1, freqs2, start):
  f1 = get_freqs(freq, N, file, freqs1, start)
  f2 = get_freqs(freq, N, file, freqs2, start)
  map_dict = {697: {1209: 1, 1336: 2, 1477: 3, 1633: 'A'}, 770: {1209: 4, 1336: 5, 1477: 6, 1633: 'B'}, 852: {1209: 7, 1336: 8, 1477: 9, 1633: 'C'}, 941: {1209: '*', 1336: 0, 1477: '#', 1633: 'D'}}
  return str(map_dict[f1][f2])

print(get_key(8000, 205, 'touchtone_1.wav', [697, 770, 852, 941], [1209, 1336, 1477, 1633], 0))
print(get_key(8000, 205, 'touchtone_8.wav', [697, 770, 852, 941], [1209, 1336, 1477, 1633], 0))
print(get_key(8000, 205, 'touchtone_pound.wav', [697, 770, 852, 941], [1209, 1336, 1477, 1633], 0))
```

```python
streak = 0
digits = []
current = -1
for start in range(int(len(wavfile.read('/content/gdrive/MyDrive/Cornell-Tech-Github/Digital-Signal-Processing/soundfiles/mystery.wav')[1]) / 205)):
  next = get_key(8000, 205, 'mystery.wav', [697, 770, 852, 941], [1209, 1336, 1477, 1633], 205 * start)
  if current == next:
    streak += 1
    if streak >= 4:
      digits.append(current)
      streak = 0
  else:
    streak = 0
  current = next
print(' '.join(digits))
```
